```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import yfinance as yf 
from functools import reduce
from pypfopt import risk_models
```

```python
stocks = pd.read_csv('niftylist.csv')
stocks
```

```python
for symbol in stocks["Symbol"]: 
    failed = []
    name = symbol + ".NS"
    try : 
        data = yf.download(name , start = "2022-01-01" , end = "2023-8-30" , interval = "1d")["Close"]
        data.to_csv("data/" + symbol + ".csv")
    except : 
        failed.append(name)
        continue  
```

```python
dataset = pd.DataFrame()
dfs = []
for symbol in stocks["Symbol"]: 
    temp = pd.read_csv("data/" +symbol + ".csv")
    dataset[symbol] = temp["Adj Close"]

dataset
    # dataset.append(temp)
```

```python
returns = dataset.pct_change()
returns
```

```python
cov_matrix_annual = returns.cov() * 252 
cov_matrix_annual
```

```python
weights = np.arange(51.0)
weights.fill(1/51)
weights
```

```python
port_variance = np.dot(weights.T , np.dot(cov_matrix_annual , weights))
port_variance
```

```python
port_volatility = np.sqrt(port_variance)
port_volatility
```

```python
SimpleAnnualReturn = np.sum(returns.mean() * weights) * 252 
SimpleAnnualReturn
```

```python
str(round(SimpleAnnualReturn , 2) * 100) + "%"
```

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models 
from pypfopt import expected_returns 

mu = expected_returns.mean_historical_return(dataset)
S = risk_models.sample_cov(dataset)

ef = EfficientFrontier(mu , S)
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
print(cleaned_weights)
ef.portfolio_performance(verbose = True)
```

```python
sum = 0 
for key,values  in cleaned_weights.items() : 
    if values != 0 : 
        sum += values 
        print(key,values)
print(sum)
```
